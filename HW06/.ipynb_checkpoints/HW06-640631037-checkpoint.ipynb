1) Which region sold the largest amount of avocado?

In this region, where the biggest lot of sold avocado came form (4046, 4225, 4770)?

```python
data.groupby(['region']).max(['Total Volume']).sort_values(['Total Volume'], ascending = False).head(1)
```

**Answer**

1. The region sold the largest amount is TotalUS sold 62,505,646.52.
2. The biggest came form lot 4046 are sold by 22,743,616.17.

2) Which region sold the smallest amount of avocado?

In this region, where the biggest lot of sold avocado came from (4046, 4225, 4770)?

```python
data.groupby(['region']).min(['Total Volume']).sort_values(['Total Volume'], ascending = True).head(1)
```

**Answer**
1. The region sold the smallest amount is MiamiFtLauderdale sold by 84.56.
2. The biggest lot came from 4046 are sold 1.31.

3) Which region sold the highest price of avocado in average?

```python
data.groupby(['region']).max(['AveragePrice']).sort_values(['AveragePrice'], ascending = False).head(1)
```

**Answer**
1. The region are sold highest average price is SanFrancisco sold by 3.25.

4) Find the total amount of income ( Avg_Price * Total_Volume ) of each region.

```python
income = data['AveragePrice'] * data['Total Volume']
data['income'] = income
data.head()
```

5) Let AVOCADO Average Weight: 4046 => 4 ounces, 4225 => 9 ounces, 4770 => 12 ounces

```python
Average_Weight_4046 = data['4046'] / 4
Average_Weight_4225 = data['4225'] / 9
Average_Weight_4770 = data['4770'] / 12
data['Average_Weight_4046'] = Average_Weight_4046
data['Average_Weight_4225'] = Average_Weight_4225
data['Average_Weight_4770'] = Average_Weight_4770
data.head()
```

6) Normally, the customers buy the avocados by unit or in a bags ?

```python
Buy_Unit = data['4046'] + data['4225'] + data['4770']
data['Buy_Unit'] = Buy_Unit
data[['Buy_Unit', 'Total Bags']].sum()
```

**Answer**
1. Customers buy avocados by unit 11,150,190,000 more than bags 4,373,176,000.
